```python
%matplotlib inline
 
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev

# # Create the plot
plt.plot(x_axis, y_axis)
plt.show()
# # Create the plot with ax.plt()
# fig, ax = plt.subplots()
# ax.plot(x_axis, y_axis)
```

## Line Chart 

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Create using the object-oriented
# Create the plot with ax.plt()
fig, ax = plt.subplots()

# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label ="Boston")

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a legend for the city of Boston.
ax.legend()

# Add grid lines.
ax.grid()

ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

plt.show()
```

## Bar Chart -Vertical

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Object oriented.
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
# # Create labels for the x and y axes.
# plt.xlabel("Date")
# plt.ylabel("Fare($)")
# # Create a title.
# plt.title("PyBer Fare by Month")
# # Add the legend.
# plt.legend()
```

## Bar Chart - Horizontal

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.xticks(np.arange(0, 51, step=5.0))

plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Object oriented.
# Create the plot with ax.plt()
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label='Chicago')

#set ticks
# ax.set_xticks(np.arange(0, 51, step=5.0)
ax.xaxis.set_ticks(np.arange(0, 51, step=5.0))


# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))

ax.invert_yaxis()

# Add the legend.
ax.legend()

# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Create a title.
ax.set_title("PyBer Fare by Month")
plt.show()
```

## Scatter Plots and Bubble Charts

```python
#Method 1 
plt.plot(y_axis,x_axis, 'o',color ='r', label='Chicago')

plt.gca().invert_yaxis()

plt.xlim(0,50)

plt.title('PyBer Fare By Month')
plt.xlabel('Fare($)')
plt.ylabel("Date")
plt.legend()

plt.show()
```

```python
#Method 2  
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Method 2
new_list = [data*3 for data in y_axis]
plt.scatter(x_axis, y_axis, s=new_list)
plt.show()
```

```python
#Method 2 Object Oriented
fig, ax=plt.subplots()
new_list = [data*10 for data in y_axis]
ax.scatter(y_axis,x_axis,  s=new_list, alpha=0.8, color='skyblue', edgecolor='black', linewidth = 2, label="Boston")
ax.invert_yaxis()
ax.legend()

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Create a title.
ax.set_title("PyBer Fare by Month")

ax.set_xlim(0,50)
plt.show()
```

## Pie Chart

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        labels=x_axis,
        explode=explode_values,
        autopct="%.1f%%") 
plt.show()
```

```python
#Object oriented 
fig, ax=plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,
        labels=x_axis,
        explode=explode_values, 
        startangle = 90, 
       shadow = True,
        autopct="%.1f%%",
      counterclock=False,
       colors =colors
    )

# plt.pie(y_axis)
plt.show()
```
